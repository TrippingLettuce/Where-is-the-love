```python
df = pd.read_csv('/home/lettuce/MyCode/pandasproject/rap_end.csv')

# Load the English model in Spacy
nlp = spacy.load('en_core_web_sm')

# Define a function to check if a word is a noun, verb, or adjective
def is_noun_verb_adj(word):
    pos = nlp(word)[0].pos_
    return pos in ['NOUN', 'VERB', 'ADJ']

# Iterate over columns and drop those not meeting the criteria
for col in df.columns:
    if not is_noun_verb_adj(col):
        df = df.drop(columns=[col])
        
df.head()
```
